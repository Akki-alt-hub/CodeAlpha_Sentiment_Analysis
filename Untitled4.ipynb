```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from the uploaded CSV file
df = pd.read_csv("/content/sample_data/Stress_Dataset.csv")

# --- Sentiment Analysis Code ---

# The sentiment analysis is based on the keywords in the 'Which type of stress do you primarily experience?' column.
# The unique values in this column are:
# - 'Eustress (Positive Stress) - Stress that motivates and enhances performance.'
# - 'Distress (Negative Stress) - Stress that causes anxiety and impairs well-being.'
# - 'No Stress - Currently experiencing minimal to no stress.'

# Define a function to classify the sentiment based on these keywords.
def get_sentiment(text):
    text_lower = str(text).lower()
    if 'positive stress' in text_lower:
        return 'Positive'
    elif 'negative stress' in text_lower:
        return 'Negative'
    else:
        # Classify 'No Stress' and any other values as Neutral
        return 'Neutral'

# Apply the function to create a new 'sentiment' column
df['sentiment'] = df['Which type of stress do you primarily experience?'].apply(get_sentiment)

# Count the occurrences of each sentiment and display the results
sentiment_counts = df['sentiment'].value_counts()
print("Sentiment Analysis Results:\n", sentiment_counts)

# --- Visualization Code (Pie Chart) ---

# Create a pie chart to visualize the sentiment distribution
plt.figure(figsize=(8, 8))
plt.pie(sentiment_counts, labels=sentiment_counts.index, autopct='%1.1f%%', startangle=140, colors=['lightgreen', 'lightcoral', 'lightblue'])
plt.title('Sentiment Distribution of Stress Types', fontsize=16)
plt.ylabel('')
plt.show()

# You can also save the plot to a file
plt.savefig('sentiment_distribution.png')
```
